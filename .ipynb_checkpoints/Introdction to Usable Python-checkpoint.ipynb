## 1.2 Speed
Python is undoubtedly slower than compiled languages, but it can be sped up a lot by using the compiled functions available in numpy. *Addition: It is also possible to compile certain algorithms with packages like Cython and Numba.*

```python
import numpy as np # load numpy rand() function into memory
import random # load pure python random() function into memory
import math # load pure python sqrt function into memory

def python_pi_example(N):
    N_under = 0 # Initiate counter variable
    
    # Perform N loop iterations
    for i in range(N): 
        # Generate a random x & y coordinate using random() from random package
        x = random.random()
        y = random.random() 
        
        # If sqrt(x**2 + y**2) <= 1 then add one to the counter variable
        if math.sqrt(x**2 + y**2) <= 1:
            N_under += 1
    
    return(4 * N_under / N) # Return pi


def python_small_pi_example(N):
    xs = [random.random() for i in range(N)]
    ys = [random.random() for i in range(N)]
    N_under = sum(1 for x, y in zip(xs, ys) if math.sqrt(x**2 + y**2) <= 1)
    return( 4 * N_under / N)


def numpy_pi_example(N):
    # Generate a 2xN array of random numbers
    x, y = np.random.rand(2, N)
    
    # Count the number of these for which np.sqrt(x**2 + y**2) <= 1
    N_under = (np.sqrt(x**2 + y**2) <= 1).sum()
    
    return(4 * N_under/ N) # Return pi


# Specify the number of random coordinates to generate
N = int(1e5)

# Run python function
print("Python Function")
print("Python pi = {:.3f}".format(python_pi_example(N)))
%timeit python_pi_example(N) # Time the pure python function using the ipython %timeit magic

# Run smaller python function
print("\nSmaller Python Function")
print("Smaller Python pi = {:.3f}".format(python_small_pi_example(N)))
%timeit python_small_pi_example(N) # Time the pure python function using the ipython %timeit magic

# Run numpy function
print("\nNumpy Function")
print("Numpy pi = {:.3f}".format(numpy_pi_example(N)))
%timeit numpy_pi_example(N) # Time the numpy function using the ipython %timeit magic
```

### Additional Example: Numba Compiled Function

```python
# Additional section about numba for compiling functions easily.
from numba import jit

# Use the @jit operator to mark a function to be compiled
@jit
def jitted_python_pi_example(N):
    N_under = 0 # Initiate counter variable
    
    # Perform N loop iterations
    for i in range(N): 
        # Generate a random x & y coordinate using random() from random package
        x = random.random()
        y = random.random() 
        
        # If sqrt(x**2 + y**2) <= 1 then add one to the counter variable
        if math.sqrt(x**2 + y**2) <= 1:
            N_under += 1
    
    return(4 * N_under / N) # Return pi

# Run python function
print("Jitted Python Function")
print("Jitted Python pi = {:.3f}".format(jitted_python_pi_example(N)))
%timeit jitted_python_pi_example(N) # Time the pure python function using the ipython %timeit magic
```

# 2 - Plotting
There are lots of plotting libraries. The most popular is certainly matplotlib which is the package I use day-to-day. However since notebooks have become popular packages which offer browser-based interactive graphs have become available. One such package is called bokeh. This allows the user to zoom in etc inside the notebook. Here's an example of a monte carlo normal distribution integration.

```python
def norm(x):
    mean = 0 
    stdev = 1
    return((1/(np.sqrt(2*np.pi)*stdev)) * np.exp(-(x-mean)**2 / 2 / stdev**2))

# Make some limits
x_lim = 3
y_lim = norm(0)

# prepare some data
x = np.arange(-x_lim, x_lim + 0.1, 0.1)
y = norm(x)

# Print the result of a trapezium rule integration for comparison.
print("Trapezium Rule Integration: {:.4f}".format(np.trapz(y, x, dx = x[1] - x[0])))


N = 1e5

# Define some random variables 
xs,ys = np.random.rand(2,int(N))
xs = xs * (2 * x_lim) - x_lim # Width of 2 centred on zero
ys = ys * y_lim # from 0 to max(norm))

# Find the sets of coordinates that are under the curve
under_curve = ys < norm(xs)

# Count the number of points under the curve
N_under = under_curve.sum()

# Do the integration
print("Monte Carlo Integration: {:.4f}".format(2*x_lim*y_lim*N_under/N))

# Make an array of points under the curve
points_under_curve = np.array([xs[under_curve], ys[under_curve]])

# Make an array of points over the curve
points_over_curve = np.array([xs[under_curve == False], ys[under_curve == False]])
```

## 2.2 Matplotlib Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", font_scale=1.7)

fig = plt.figure(figsize=(15, 7))
plt.hist(points_under_curve[0], bins=50, normed=True, label="Inside Histogram")
plt.plot(points_under_curve[0][::80], points_under_curve[1][::80], 'o', label="Under Curve")
plt.plot(points_over_curve[0][::80], points_over_curve[1][::80], 'o', label="Over Curve")
plt.plot(x,y, '-', label="Normal Function")
plt.xlim(-x_lim, x_lim)
plt.show()
```

## 2.3 Bokeh Plots (interactive browser based plots)

```python
from bokeh.plotting import figure, show, output_notebook
output_notebook()

# create a new plot with a title and axis labels
p = figure(x_axis_label='x', y_axis_label='y', width=900, height=500)

# Add the histogram
hist, edges = np.histogram(points_under_curve[0], density=True, bins=50)
p.quad(top = hist, bottom = 0, left=edges[:-1], right=edges[1:],
       fill_color="orange", line_color="black", legend="Inside Histogram")

# Plot the random values inside the curve
p.circle(points_under_curve[0][::80], points_under_curve[1][::80],
         fill_color="green", line_color=None, legend="Inside Curve")

# Plot the random values that are outside the curve
p.circle(points_over_curve[0][::80], points_over_curve[1][::80],
         fill_color="grey", line_color=None, legend="Outside Curve")

# Plot a line of the function.
p.line(x, norm(x), legend="Plotted Value")

# show the results
show(p)
```

## 3.3 Data Analysis
Pandas isn't just for storing & manipulating data with handy data frames. It contains a lot of stuff for data analysis too. In fact I have only ever used a fraction of the functionality that pandas offers, it really is quite vast. Here are some general tools that I thought might be helpful.

### 3.3.1 Quick Gaussian Analysis

```python
#Percent Change
pct_chg_df = pd.DataFrame({'a' : [0, 25, 50, 75, 100, 125, 150, 175], 'b' : [1, 26, 51, 76, 101, 126, 151, 176]})
print(pct_chg_df)
print("\n")

# Percent Change
print("Pecent Change: \n{}".format(pct_chg_df.pct_change()))
print("\n")

# Mean
print("Mean: \n{}".format(pct_chg_df.mean()))
print("\n")

# Standard Deviation
print("Standard Deviation: \n{}".format(pct_chg_df.std()))
print("\n")

# Sum
print("Sum: \n{}".format(pct_chg_df.sum()))
print("\n")
```

### 3.3.2 More Analysis using Additional Packages

```python
# Standard Error on Mean = std / sqrt(N)
from scipy import stats
import scikits.bootstrap as boot

# Make some random gaussian shaped data around 3. Not very much data.
d = np.random.randn(100) + 3
plt.hist(d, bins=50)

print("Standard Error on mean: \n{:.2f} ± {:.2f} \n".format(np.mean(d), stats.sem(d)))

# Bootstrapping 95% confidence levels
print("Bootstrapped 95% mean confidence levels: \n{}".format(
        boot.ci(d, np.mean, alpha = 0.05, n_samples=10000)
    ))
```

### 3.3.3 Fitting

We'll now do a least squares fit to some data. I'll generate some random x values, and use a quadratic for y. Specifically I'll use 

$$
y = -5.45x^3 + 3.62x^2 + 2.733x + 5.17,
$$

and we'll see how well it goes.

```python
# Generate some random x-data
x = np.random.rand(30)*10 - 5

# Put this random data in a pandas dataFrame along with some y-values
ols_df = pd.DataFrame({'x': x, 'y': -5.45*x**3 + 3.62*x**2 + 2.733*x + 5.17})
```

### Numpy Polyfit

```python
# Do a numpy polyfit on the data in the data frame
ols_df_pf = np.polyfit(ols_df['x'], ols_df['y'], 3)

# Make a range of x values for which o plot
xf = np.arange(min(x), max(x)+0.01, 0.01)

# Get some polyfitted y-values for these x-values
polyfit_y = np.polyval(ols_df_pf, xf)

# Print the results
print("Numpy Polyfit Results: {}".format(ols_df_pf))
```

### Statsmodels OLS

```python
# Least Square Fitting using the statsmodels.formula.api works really well with the DataFrames
import statsmodels.formula.api as smf
from numba import vectorize

def ols_eval(y_fit, x):
    y = y_fit.params[0]
    n = len(y_fit.params) - 1
    for i in y_fit.params[1:]:
        y += i*x**n
        n -= 1
        
    return y

# Perform a fit of the y column with the formula y = pow(x,2) + x. pow(x,2) == x**2
y_fit = smf.ols(formula="y ~ pow(x,3) + pow(x,2) + x", data = ols_df).fit()

# Get some y values for a set of x's
ols_y = ols_eval(y_fit, xf)

# Print the summary & statistics of the OLS fit
print(y_fit.summary())
```

### Results Plotted

```python
# Plot the raw (x,y) data and the fitted data
plt.figure(figsize=(16,5))
plt.plot(ols_df['x'], ols_df['y'], 'o', label="Data")
plt.plot(xf, ols_y, '-', label="Statsmodels OLS Fit")
plt.plot(xf, polyfit_y, '--', label="Numpy Polyfit")
plt.legend(loc=0)
plt.show()
```

## 1. Fibonacci Sequence
Make a code to return the first $n$ fibonacci numbers.
There are a few possible solutions below.

```python
def py_fib(n):
    fib_numbers = [1, 1]
    [fib_numbers.append(fib_numbers[-1] + fib_numbers[-2]) for i in range(n-2)]
    return fib_numbers[0:n]

@jit
def jit_fib(n):
    fib_numbers = [1, 1]
    for i in range(n-2):
        fib_numbers.append(fib_numbers[-1] + fib_numbers[-2])
    return fib_numbers


def npfib(n):
    fib_numbers = np.array([1, 1])
    [np.append(fib_numbers, (fib_numbers[-1] + fib_numbers[-2])) for i in range(n-2)]
    return fib_numbers[0:n]

%timeit py_fib(10)
print("\n")
%timeit jit_fib(10)
print("\n")
%timeit npfib(10)
```

# 5 - Symbolic Maths
The packages we've talked about before are great for numerical calculation, but can't be used for symbolic manipulation. You might be tempted to pick up a different package for this. No need, sympy as you covered.

```python
import sympy as sm
from IPython.display import display, Latex
sm.init_printing()

x = sm.Symbol('x')

print("Differente x**2")
display(sm.diff(x**2))
print("\n")

print("Integrate x**2")
display(sm.integrate(x**2))
print("\n\n")

print("Solving f(x) = x**2 + x + 1")
display(sm.solve(x**2 + x + 1, x))
print("\n")

print("Substituting f(5)")
f = sm.lambdify(x, x**2 + x + 1)
print(f(5))
```
